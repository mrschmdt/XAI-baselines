```python
from OpenXAI.openxai.dataloader import return_loaders, TabularDataLoader
from umap import UMAP
from scipy.spatial.distance import pdist, squareform
import matplotlib.pyplot as plt


gauss_params = {
    'n_samples': 2500,
    'dim': 20,
    'n_clusters': 2,
    'distance_to_center': 5,
    'test_size': 0.25,
    'upper_weight': 1,
    'lower_weight': -1,
    'seed': 564,
    'sigma': None,
    'sparsity': 0.75
}

dataset = TabularDataLoader(
    path="Synthetic",
    filename="train",
    gauss_params=gauss_params,
    label="y",
    scale="minmax")

X = dataset.data
Y = dataset.cluster_idx

D = pdist(X, metric='euclidean')
D = squareform(D)



pass
```

```python
reducer = UMAP(n_components=2, metric="precomputed")

umap = reducer.fit_transform(D)
```

```python
import numpy as np

colors = np.array(['blue','orange'])
plt.scatter(umap[:,0],umap[:,1], c=colors[Y])
```
